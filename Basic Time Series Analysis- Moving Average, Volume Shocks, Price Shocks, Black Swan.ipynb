```python
#from my research on the internet it is a simple average of a price over a period 
#so, here I interpret the 4 week moving average for closing price
#as the average of closing prices over 4 weeks from each point

def simple_moving_average(data,week_val):
    data.index = pd.to_datetime(data.index)
    period = timedelta(weeks = week_val)
    moving_avg = []
    last_index = data.index[len(data)-1]
    last_ind_iter = last_index - period 
    iter_val = len(data.loc[data.index[0]:last_ind_iter,:]) + 1
    for i in range(0,iter_val):
        # The last fiew values won't be meaningful without the iter_value
        end = data.index[i] + period
        subset = data.loc[data.index[i]:end,:]
        avg = subset["Close"].mean()
        moving_avg.append(avg)
    return moving_avg
    
```

```python
# using business day data

def volume_shocks(data_2):
    data = data_2.copy(deep = True)
    data["% change"] = float('0')
    
    for i in range(0,(len(data)-1)):
        data["% change"].iloc[i+1] = (((data["Volume"].iloc[i+1] - data["Volume"].iloc[i])/data["Volume"].iloc[i])*100)
    data["abs_change"] = np.absolute(data["% change"])
    shock = []
    direction = []
    
    for index,row in data.iterrows():
        if row["abs_change"] >= 10:
            shock.append(int(1))
        else:
            shock.append(int(0))
    for index,row in data.iterrows():    
        if (row["% change"] > 0 and row["abs_change"] >= 10):
            direction.append(int(1))
        elif (row["% change"] < 0 and row["abs_change"] >= 10):
            direction.append(int(0))
        else:
            direction.append(np.nan)
          
        
        
    data["shock"] = pd.Series(shock,index = data.index)
    data["direction"] = pd.Series(direction,index = data.index)
    
    data_2["shock"] = data["shock"].copy(deep=True)
    data_2["direction"] = data["direction"].copy(deep = True)
    
    
    return data_2    
        
    
```

```python
# using business day data

def price_shocks(data_2):
    data = data_2.copy(deep = True)
    data["% change"] = float('0')
    
    for i in range(0,(len(data)-1)):
        data["% change"].iloc[i+1] = (((data["Close"].iloc[i+1] - data["Close"].iloc[i])/data["Close"].iloc[i])*100)
    data["abs_change"] = np.absolute(data["% change"])
    shock = []
    direction = []
    
    for index,row in data.iterrows():
        if row["abs_change"] >= 2:
            shock.append(int(1))
        else:
            shock.append(int(0))
    for index,row in data.iterrows():    
        if (row["% change"] > 0 and row["abs_change"] >= 2):
            direction.append(int(1))
        elif (row["% change"] < 0 and row["abs_change"] >= 2):
            direction.append(int(0))
        else:
            direction.append(np.nan)
          
        
        
    data["shock"] = pd.Series(shock,index = data.index)
    data["direction"] = pd.Series(direction,index = data.index)
    
    data_2["price_shock"] = data["shock"].copy(deep=True)
    data_2["price_direction"] = data["direction"].copy(deep = True)
    
    
    return data_2    
        
    
```

```python
# using business day data
#the instructions said two percent for black swan, but that was
#the criteria for a shock
# A pricing black swan is like a stock market crash event, so taking the 
#price % change as 20-30% for INFY, TCS and the NiftyIT would be
# safe to be considered as a black swan, I am letting it be an input,


def black_swan(data_2, change):
    data = data_2.copy(deep = True)
    data["% change"] = float('0')
    
    for i in range(0,(len(data)-1)):
        data["% change"].iloc[i+1] = (((data["Close"].iloc[i+1] - data["Close"].iloc[i])/data["Close"].iloc[i])*100)
    data["abs_change"] = np.absolute(data["% change"])
    swan = []
    direction = []
    
    for index,row in data.iterrows():
        if row["abs_change"] >= change:
            swan.append(int(1))
        else:
            swan.append(int(0))
    for index,row in data.iterrows():    
        if (row["% change"] > 0 and row["abs_change"] >= change):
            direction.append(int(1))
        elif (row["% change"] < 0 and row["abs_change"] >= change):
            direction.append(int(0))
        else:
            direction.append(np.nan)
          
        
        
    data["swan"] = pd.Series(swan,index = data.index)
    data["direction"] = pd.Series(direction,index = data.index)
    
    data_2["price_black_swan"] = data["swan"].copy(deep=True)
    data_2["black_swan_direction"] = data["direction"].copy(deep = True)
    
    
    return data_2    
        
    
```

```python
infosys = black_swan(infosys,20)
```

```python
tcs = black_swan(tcs,20)
nifty_it = black_swan(nifty_it,20)
```

```python
# nifty did not either
# price shock without volume shock

def price_no_volume(data2):
    data = data2.copy(deep=True)
    p_no_v =[]
    for i in range(0,len(data)):
        if (data["price_shock"].iloc[i] == 1 and data["shock"].iloc[i] == 0):
            p_no_v.append(int(1))
        else:
            p_no_v.append(int(0))
    data["p_no_v"] = pd.Series(p_no_v,index = data.index)
    data2["priceshock_wo_volumeshock"] = data["p_no_v"].copy(deep=True)
    return data2
    
```
